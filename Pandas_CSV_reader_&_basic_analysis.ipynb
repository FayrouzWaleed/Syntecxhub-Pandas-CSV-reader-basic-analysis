## Pandas

```python
import pandas as pd
```

```python
people = pd.read_csv('./people-100.csv')
```

## Encoding Dataset (to get stats)

```python
people["Sex_encoded"] = people["Sex"].map({
    "Male": 1,
    "Female": 0
})
```

```python
people.head()
```

```python
people["JobTitle_encoded"] = people["Job Title"].astype("category").cat.codes
```

```python
people.head()
```

```python
people["Date of birth"] = pd.to_datetime(people["Date of birth"])

today = pd.Timestamp.today()
people["Age"] = (today - people["Date of birth"]).dt.days // 365
```

```python
people.head()
```

```python
people[["Sex_encoded", "JobTitle_encoded", "Age"]].describe()
```

```python
people[["JobTitle_encoded", "Age"]].median()
```
